4.4 Look into MultiIndex Layers (Stack Functions)

```python
df = pd.read_csv("worldstats.csv", index_col = ["country", "year"])
```

```python
#shifting the dataframe 將資料變成一維（column下放成為index的一層layer)
#return Series
df = df.stack().to_frame()
df.head(3)
```

```python
#反操作，將index layer shift back default level = -1 ，可以選擇哪一層layer
df = df.unstack(2).head(3)
df.head(3)
```

```python
df = df.unstack(0)
df.head(3)
```

```python
#shilt multiple layers
df.unstack(level = [1, 0]).head(3)
```

## 5 GroupBy

```python
df = pd.read_csv("fortune1000.csv")
df
```

5.1 Generate a DataFrameGroupBy Object and Some Available Methods on such

```python
#generate with one column
sector = df.groupby("Sector")
#with multiple columns
df.groupby(["Sector", "Location"])
```

```python
#length
len(sector)
#size
sector.size()
#similar result with .value_counts()
df["Sector"].value_counts()
```

```python
#only the first occurence of each group value
sector.first()
#only the first occurence of each group value
sector.last()
```

```python
#property, return a dictionary of the form {group_value : [index with given group_value], ...}
sector.groups
```

```python
#filtering the subset with desired value
sector.get_group("Energy")
#previous approach
df[df["Sector"] == "Energy"]
```

```python
#mathematcal operation
#min, chosen based the column next to index, in this case "Rank"
sector.min()
#sum, every available columns
sector.sum()
```

```python
# perform operation on selected group and column
sector.get_group("Energy")["Profits"].sum()
#on given column
sector["Profits"].sum()
```

```python
#similar method with advanced settings ( .agg() )
#feed a dictionary
sector.agg({"Revenue": "sum", "Profits": "mean","Rank":"min"} )
#feed a list, perform operations on each column
sector.agg(["size", "sum", "mean"])
#combination
sector.agg({"Profits": ["mean", "size"], "Revenue": "sum"})
```

## 6 Merging, Joining, and Concatenating

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
```

6.1 pd.concat() and .append() (vertival)

```python
#(vertical) concatenation
pd.concat([week1, week2])
#generate new index
pd.concat([week1, week2], ignore_index = True)
```

```python
#dataframe_name.append()
week1.append(week2, ignore_index = True)
```

6.2 Merge

```python
#inner join 交集 
week1.merge(week2, how = 'inner')
#compare by single coulmn (shared name in this case)
#if a value appears twice in "one" of the dataframe, the concatenated datagrame would have repeated rows
week1.merge(week2, how = 'inner', on = "Customer ID")
#compare by multiple columns
week1.merge(week2, how = 'inner', on = ["Customer ID", "Food ID"])
#change column suffix (needed when column name is shared)
week1.merge(week2, how = 'inner', on = "Customer ID", suffixes = ["-week 1", "-week2"])
```

```python
#outer join 去掉交集
#with indicator
df = week1.merge(week2, how = 'outer', on = "Customer ID", indicator = True)
#info on indicator
df["_merge"].value_counts()
```

```python
week2.rename(columns = {"Customer ID":"Customer ID", "Food ID": "ID"}, inplace = True)
week2.head(3)
```
